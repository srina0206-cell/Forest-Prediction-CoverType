```python
import pandas as pd

def handle_outliers_iqr(df, outlier_columns):

    df = df.copy()

    for col in outlier_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR



        df[col] = df[col].apply(
            lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
            )

    return df
```

# **Feature Engineering**

```python
#hydrology ratio
df['Hydrology_Ratio'] = df['Vertical_Distance_To_Hydrology'] / (df['Horizontal_Distance_To_Hydrology'] + 1)
```

```python
# fire and road distance
df['Fire_to_Road'] = df['Horizontal_Distance_To_Fire_Points'] / (df['Horizontal_Distance_To_Roadways'] + 1)
```

```python
#hillshade mean
df['Hillshade_mean'] = (df['Hillshade_9am'] + df['Hillshade_Noon'] + df['Hillshade_3pm']) / 3
```

```python
#elevation slope
df['Elevation_Slope'] = df['Elevation'] * df['Slope']
```

```python
#elevation hydrology
df['Elevation_Hydrology'] = df['Elevation'] / (df['Horizontal_Distance_To_Hydrology'] + 1)
```

```python
df.isnull().sum()
df.skew(numeric_only=True)
```

```python
import numpy as np

exclude_cols = ['Wilderness_Area', 'Soil_Type']

for col in df.columns:

    if col in exclude_cols:
        continue

    if df[col].dtype in ['int64', 'float64']:

        skew = df[col].skew()

        if skew > 1:
            df[col] = np.log1p(df[col].clip(lower=0))

        elif skew < -1:
            df[col] = np.log1p((df[col].max() + 1) - df[col])
```

```python

df['Fire_to_Road'] = np.log1p(df['Fire_to_Road'])
df['Elevation_Hydrology'] = np.log1p(df['Elevation_Hydrology'])


max_val = df['Hillshade_mean'].max()
df['Hillshade_mean'] = np.log1p(max_val - df['Hillshade_mean'])
```

```python
%%writefile Forest.py
import streamlit as st
import numpy as np
import pickle


# Load Model

model = pickle.load(open("random_forest_model.pkl", "rb"))


# Label Mapping (IMPORTANT)

cover_type_mapping = {
    0: "Aspen",
    1: "Cottonwood/Willow",
    2: "Douglas-fir",
    3: "Krummholz",
    4: "Lodgepole Pine",
    5: "Ponderosa Pine",
    6: "Spruce/Fir"
}


# Page Config

st.set_page_config(page_title="Forest Cover Prediction", page_icon="🌲")




# Title

st.markdown("<h1 style='text-align:center;'>🌲 Forest Cover Type Prediction App</h1>", unsafe_allow_html=True)

st.write("Enter the values for the following features:")


# INPUTS


elevation = st.number_input("Elevation", 0, 5000, 2000)

horizontal_distance_roadways = st.number_input(
    "Horizontal Distance To Roadways", 0, 5000, 500
)

horizontal_distance_fire = st.number_input(
    "Horizontal Distance To Fire Points", 0, 5000, 600
)

horizontal_distance_hydrology = st.number_input(
    "Horizontal Distance To Hydrology", 0, 5000, 300
)

vertical_distance_hydrology = st.number_input(
    "Vertical Distance To Hydrology", -500, 500, 0
)


hillshade_9am = st.slider("Hillshade 9am", 0, 255, 150)
aspect = st.slider("Aspect", 0, 360, 90)
hillshade_3pm = st.slider("Hillshade 3pm", 0, 255, 100)
hillshade_noon = st.slider("Hillshade Noon", 0, 255, 200)
slope = st.slider("Slope", 0, 90, 10)

# Categorical
wilderness_area = st.selectbox("Wilderness Area", [0, 1, 2, 3])
soil_type = st.selectbox("Soil Type", list(range(40)))


# AUTO FEATURE ENGINEERING


auto_hydrology_ratio = horizontal_distance_hydrology / (vertical_distance_hydrology + 1)
auto_fire_to_road = horizontal_distance_fire / (horizontal_distance_roadways + 1)
auto_hillshade_mean = (hillshade_9am + hillshade_noon + hillshade_3pm) / 3
auto_elevation_slope = elevation * slope
auto_elevation_hydrology = elevation - horizontal_distance_hydrology


hydrology_ratio = st.number_input("Hydrology Ratio", value=float(auto_hydrology_ratio))
fire_to_road = st.number_input("Fire to Road Ratio", value=float(auto_fire_to_road))
hillshade_mean = st.number_input("Hillshade Mean", value=float(auto_hillshade_mean))
elevation_slope = st.number_input("Elevation * Slope", value=float(auto_elevation_slope))
elevation_hydrology = st.number_input("Elevation - Hydrology", value=float(auto_elevation_hydrology))


# Prediction

if st.button("Predict Forest Cover Type"):

    input_data = np.array([[
        elevation,
        aspect,
        slope,
        horizontal_distance_hydrology,
        vertical_distance_hydrology,
        horizontal_distance_roadways,
        hillshade_9am,
        hillshade_noon,
        hillshade_3pm,
        horizontal_distance_fire,
        wilderness_area,
        soil_type,
        hydrology_ratio,
        fire_to_road,
        hillshade_mean,
        elevation_slope,
        elevation_hydrology
    ]])

    try:
        prediction = model.predict(input_data)[0]
        label = cover_type_mapping.get(prediction, "Unknown")

        st.success(f" Predicted Forest Cover Type: {label}")

    except Exception as e:
        st.error(f" Error: {e}")

```

# **Hypertuning**

```python
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    "n_estimators": [100, 200, 300, 400],
    "max_depth": [10, 20, 30, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
```

```python
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
random_search.fit(x_train, y_train)
```

```python
best_model = random_search.best_estimator_

print("Best Parameters:", random_search.best_params_)
```

```python
with open("best_random_forest.pkl", "wb") as f:
    pickle.dump(best_model, f)
```
